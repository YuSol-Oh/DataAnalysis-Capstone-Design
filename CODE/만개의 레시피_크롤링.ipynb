```python
import urllib.request
from bs4 import BeautifulSoup as bs
import pandas as pd
import webbrowser
```

```python
def url_func(n, m) :
    num_range = range(n, m)
    url = "https://www.10000recipe.com/recipe/list.html?cat4=63&order=reco&page="
    url_list = []
    
    for i in num_range :
        req = urllib.request.Request(url + str(i))
        code = urllib.request.urlopen(url).read()
        soup = bs(code, "html.parser")
        
        res = soup.find(class_ = 'common_sp_list_ul ea4')
        for i in res.find_all('a') :
            url_tmp = i.get('href')
            url_list.append(url_tmp)
        
    return url_list
```

```python
menu_name = []
menu_img = []
menu_summary = []
menu_info_1 = []
menu_info_2 = []
menu_info_3 = []
ingredient_list = []
recipe_list = []
num_id = 0
food_dicts = []

webbrowser.open("https://www.10000recipe.com/recipe/list.html?cat4=63&order=reco&page=")

url_list = url_func(1, 3)
for url_str in url_list :
    url = "https://www.10000recipe.com"
    url = url + url_str
    req = urllib.request.Request(url)
    code = urllib.request.urlopen(url).read()
    soup = bs(code, "html.parser")
    
    # menu name
    try : 
        res = soup.find('div', 'view2_summary st3')
        res = res.find('h3')
        menu_name.append(res.get_text())
    except(AttributeError) :
        pass
    
    # menu_img
    try : 
        res = soup.find('div', 'centeredcrop')
        res = res.find('img')
        menu_img.append(res.get('src'))
    except(AttributeError) :
        pass
    
    # menu_summary
    try :
        res = soup.find('div', 'view2_summary_in')
        menu_summary.append(res.get_text().replace('\n', '').strip())
        
    except(AttributeError) :
        pass

    # menu_info
    #res = soup.find('span', 'view2_summary_info1')
    #menu_info_1.append(res.get_text()) # n인분
    #res = soup.find('span', 'view2_summary_info2')
    #menu_info_2.append(res.get_text()) # 요리시간
    #res = soup.find('span', 'view2_summary_info3')
    #menu_info_3.append(res.get_text()) # 난이도
    
    # ingredient
    res = soup.find('div', 'ready_ingre3')
    ingredient = []
    
    try : 
        if len(res.find_all('ul')) < 2 :
            for i in res.find_all('ul') :
                ingredient = []  
                for k in i.find_all('li') :
                    ingredient.append(k.get_text().replace('\n', ' ').replace('구매', ' ').replace(' ', '')) 
                ingredient_list.append(ingredient)
        else : 
            ingredient = []
            for i in res.find_all('ul') :
                ingredient = ingredient
                for k in i.find_all('li') :
                    ingredient.append(k.get_text().replace('\n', ' ').replace('구매', ' ').replace(' ', ''))
            ingredient_list.append(ingredient)
            
    except(AttributeError) :
        pass
    
    # recipe
    try :
        res = soup.find('div', 'view_step')
        recipe = []
        for i in res.find_all('div', 'media-body') :
            text = i.get_text()
            if "     " in text :
                text = text.split("     ")[0]
            
            recipe.append(text.replace('\n', ''))
        recipe_list.append(recipe)
        recipe = []
        
    except(AttributeError) :
        pass
    
```

```python
print(menu_name)
print(ingredient_list)
print(recipe_list)
```

```python
# csv 파일로 저장하기

import pandas as pd

data = pd.DataFrame(list(zip(menu_name,ingredient_list,recipe_list)), columns=['Menu','Ingredient','Recipe'])


data.to_csv('1페이지.csv',index = False,encoding="utf-8-sig") # recipe_(분류값)_(순서)
```
